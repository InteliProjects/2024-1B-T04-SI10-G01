```python
# @title data vs qtd_leads

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['data']
  ys = series['qtd_leads']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df_atual.sort_values('data', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('data')
_ = plt.ylabel('qtd_leads')
```

*   **carregamento_pagina**: Provavelmente refere-se ao evento registrado quando uma página é carregada no navegador do usuário. Isso é útil para entender a performance da página e a frequência de acessos.

*   **interaction**: Geralmente captura qualquer tipo de interação que o usuário tem com o site, que não seja um clique, como digitar em um campo de formulário.
*   **scroll**: Este evento é registrado quando o usuário rola a página para cima ou para baixo. Ajuda a entender até que ponto da página o usuário está interessado em ler ou visualizar.
*   **video_complete**: Registra quando um vídeo é assistido até o fim. Esse dado é essencial para avaliar a eficácia e o engajamento do conteúdo de vídeo.
*   **video_start**: Semelhante ao play_video, registra cada vez que um vídeo é iniciado, o que pode incluir repetições ou novas visualizações após pausas.

# Simulação de Monte Carlo

## Tempo Médio de Sessão x Taxa de Conversão

```python
# Importa as bibliotecas necessárias
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.utils import resample
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
fig = px.scatter(
    df_wizardon,
    x="media_sessao",
    y="taxa_de_conversao",
    trendline="ols",  # Adiciona uma linha de regressão linear ordinária ('ordinary least squares')
    title="Relação entre Tempo de Sessão e Taxa de Conversão",
    labels={"Tempo Médio de Sessão (minutos)": "Tempo Médio de Sessão (min)",
            "Taxa de Conversão (%)": "Taxa de Conversão (%)"}
)
fig.update_layout(
    xaxis_title="Tempo Médio de Sessão (min)",
    yaxis_title="Taxa de Conversão (%)",
    height=500,
    width=800
)
fig.show()
```
